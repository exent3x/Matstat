### b) $n=2$


Отношение правдоподобия:

$$
\Lambda(x)=\frac{L_1}{L_0}=\frac{(\frac{e}{e-1})^2*e^{-x_1-x_2}}{1*1}\propto e^{-(x_1+x_2)}\ge C\;\Longrightarrow\;x_1+x_2\le A
$$

Критическая область $G:\;x_1+x_2\le A$.

Найдём $A$ из условия $P(G\mid H_0)=\alpha$:

$$
\iint\limits_{x_1+x_2\le A,\;x_i>0}1\cdot dx_1dx_2=\frac{A^2}{2}=\alpha\;\Longrightarrow\;A=\sqrt{2\alpha}\quad(\alpha\le0.5)
$$

Таким образом, $G:\;x_1+x_2\le\sqrt{2\alpha}$.

Мощность:

$$
W=P(G\mid H_1)=\iint\limits_{x_1+x_2\le\sqrt{2\alpha}}\left(\frac{e}{e-1}\right)^2 e^{-x_1-x_2}dx_1dx_2
$$

Вычисление:

$$
W=\left(\frac{e}{e-1}\right)^2\int_0^{\sqrt{2\alpha}} e^{-s}\cdot s\,ds
   =\frac{e^2}{(e-1)^2}\Bigl(1-(1+\sqrt{2\alpha})e^{-\sqrt{2\alpha}}\Bigr)
$$
Ошибка первого рода:$\alpha_1=\alpha$
Ошибка второго рода: $\alpha_2=1-W$.

### d) Критерий, основанный на $X_{\min}$

Критическая область: $G:\;X_{\min}\le C$.

При $H_0$: $F_0(x)=x$, функция распределения минимума:

$$
P(X_{\min}\le C\mid H_0)=1-(1-C)^n=\alpha\;\Longrightarrow\;C=1-\sqrt[n]{1-\alpha}.
$$

Мощность:

$$
W=P(G\mid H_1)=1-\bigl(1-F_1(C)\bigr)^n,
$$

где $F_1(x)=\int_0^x\frac{e}{e-1}e^{-t}dt=\frac{e}{e-1}(1-e^{-x})$.

Подставляя $C$:

$$
W=1-\left(1-\frac{e}{e-1}\bigl(1-e^{-C}\bigr)\right)^n
   =1-\left(1-\frac{e}{e-1}+\frac{e}{e-1}e^{-C}\right)^n.
$$

При $n\to\infty$: $C=1-\sqrt[n]{1-\alpha}\to0$, $e^{-C}\to1$, поэтому

$$
1-\frac{e}{e-1}+\frac{e}{e-1}e^{-C}\to 1-\frac{e}{e-1}+\frac{e}{e-1}=1,
$$

$$
C=1-(1-\alpha)^{1/n}=1-e^{\frac1n\ln(1-\alpha)}\approx -\frac1n\ln(1-\alpha)+O\Bigl(\frac1{n^2}\Bigr).
$$

Тогда $e^{-C}\approx 1+\frac1n\ln(1-\alpha)$. Подставляя:

$$
1-\frac{e}{e-1}+\frac{e}{e-1}\Bigl(1+\frac1n\ln(1-\alpha)\Bigr)=1+\frac{e}{e-1}\cdot\frac{\ln(1-\alpha)}{n}+O\Bigl(\frac1{n^2}\Bigr).
$$

Следовательно,

$$
\left(1+\frac{e\ln(1-\alpha)}{e-1}\cdot\frac1n\right)^n \xrightarrow[n\to\infty]{} \exp\left(\frac{e\ln(1-\alpha)}{e-1}\right)=1-(1-\alpha)^{\frac{e}{e-1}}.
$$

Таким образом,

$$
W\to 1-(1-\alpha)^{\frac{e}{e-1}}<1.
$$

Критерий **несостоятельный**, так как мощность не стремится к $1$ при $n\to\infty$.
