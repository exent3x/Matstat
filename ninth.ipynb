# Задача 9

Проверяется гипотеза $H_0$: распределение оценок одинаково для двух потоков (поток и оценка независимы). Данные:

| Поток | Оценка 2 | Оценка 3 | Оценка 4 | Оценка 5 | Сумма |
|-------|----------|----------|----------|----------|-------|
| 1     | 33       | 43       | 80       | 144      | 300   |
| 2     | 39       | 35       | 72       | 154      | 300   |
| Сумма | 72       | 78       | 152      | 298      | 600   |

Ожидаемые частоты при $H_0$:

$$
E_{11}=\frac{300\cdot72}{600}=36,\quad E_{12}=39,\quad E_{13}=76,\quad E_{14}=149,
$$
$$
E_{21}=36,\quad E_{22}=39,\quad E_{23}=76,\quad E_{24}=149.
$$

Статистика $\chi^2$:

$$
\chi^2_{\text{набл}} = \sum_{i,j}\frac{(n_{ij}-E_{ij})^2}{E_{ij}} = 2\cdot\left(\frac{(33-36)^2}{36}+\frac{(43-39)^2}{39}+\frac{(80-76)^2}{76}+\frac{(144-149)^2}{149}\right) = 2{,}0772.
$$

Число степеней свободы $\text{df} = (2-1)(4-1) = 3$.  
Критическое значение $\chi^2_{0{,}05}(3) \approx 7{,}815$. 
При справедливости $H_0$ статистика $\chi^2$ имеет распределение $\chi^2$ с $3$ степенями свободы. Плотность этого распределения:

$$
f_{\chi^2(3)}(t) = \frac{1}{2^{3/2}\,\Gamma(3/2)}\, t^{1/2} e^{-t/2},\quad t>0,
$$
где $\Gamma(3/2) = \frac{\sqrt{\pi}}{2}$.

P‑значение:

$$
\text{P-value} = P(\chi^2(3) \ge 2.077134) = \int_{2.077134}^{+\infty} f_{\chi^2(3)}(t)\,dt.
$$

Вычислим интеграл численно. Функция распределения $\chi^2$ с $3$ степенями свободы выражается через неполную гамма-функцию:
$$
F(x) = \frac{\gamma(3/2,\, x/2)}{\Gamma(3/2)},
$$
где $\gamma$ — нижняя неполная гамма-функция. Тогда
$$
\text{P-value} = 1 - F(2.077134).
$$
$$
F(2.077134) \approx 0.554,\quad \text{P-value} \approx 0.446.
$$


Поскольку P‑значение $0.4456 > 0.05$, гипотеза $H_0$ не отвергается на уровне значимости $0.05$.

**Вывод:** нет оснований считать распределение оценок в двух потоках различным; оба потока однородны
