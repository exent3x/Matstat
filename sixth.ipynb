Случайная величина $X$ имеет распределение Парето с плотностью  
$p(x;\theta) = \begin{cases} \dfrac{\theta-1}{x^{\theta}}, & x\ge 1,\\ 0, & x<1 \end{cases}, \qquad \theta>1.$

---

## a) Оценка максимального правдоподобия для $\theta$

Выборка $X_1,\dots,X_n$. Функция правдоподобия  
$L(\theta)=\prod_{i=1}^n\frac{\theta-1}{X_i^{\theta}} = (\theta-1)^n \prod_{i=1}^n X_i^{-\theta}.$  
Логарифмическая функция  
$\ln L(\theta)=n\ln(\theta-1)-\theta\sum_{i=1}^n\ln X_i.$  
Производная  
$\frac{\partial\ln L}{\partial\theta}=\frac{n}{\theta-1}-\sum_{i=1}^n\ln X_i.$  
Приравнивая к нулю, получаем  
$\frac{n}{\theta-1}=\sum_{i=1}^n\ln X_i \quad\Rightarrow\quad \hat\theta_{\text{МП}}=1+\frac{n}{\sum_{i=1}^n\ln X_i}.$  
Вторая производная  
$\frac{\partial^2\ln L}{\partial\theta^2}=-\frac{n}{(\theta-1)^2}<0,$  
---

## b) Асимптотический доверительный интервал для медианы

Медиана $m$ находится из условия $\int_1^m p(x;\theta)\,dx=0.5$:  
$\int_1^m (\theta-1)x^{-\theta}dx = \bigl[-x^{-(\theta-1)}\bigr]_1^m = 1-m^{-(\theta-1)}=0.5\quad\Rightarrow\quad m=2^{\frac{1}{\theta-1}}.$  
Обозначим $g(\theta)=2^{\frac{1}{\theta-1}}$. Оценка максимального правдоподобия $\hat\theta$ асимптотически нормальна:  
$\sqrt{n}(\hat\theta-\theta)\xrightarrow{d} N(0,(\theta-1)^2).$  
По дельта-методу  
$\sqrt{n}\bigl(g(\hat\theta)-g(\theta)\bigr)\xrightarrow{d} N\bigl(0,\; (g'(\theta))^2(\theta-1)^2\bigr).$  
Вычислим производную:  
$g'(\theta)=2^{\frac{1}{\theta-1}}\cdot\ln2\cdot\frac{-1}{(\theta-1)^2}= -\frac{m\ln2}{(\theta-1)^2}.$  
Таким образом, асимптотическая дисперсия $g(\hat\theta)$ равна  
$\frac{m^2(\ln2)^2}{(\theta-1)^2}.$  
Заменяя неизвестные $\theta$ и $m$ на их оценки, получаем стандартную ошибку  
$\widehat{se}=\frac{\hat m\ln2}{|\hat\theta-1|},$  
где $\hat m=2^{1/(\hat\theta-1)}$. Тогда асимптотический доверительный интервал для медианы уровня $1-\alpha$:  
$\boxed{\hat m \pm z_{1-\alpha/2}\,\frac{\hat m\ln2}{\hat\theta-1}}.$  
---
## c) Асимптотический доверительный интервал для $\theta$

Из асимптотической нормальности оценки МП  
$\sqrt{n}(\hat\theta-\theta)\xrightarrow{d} N(0,(\theta-1)^2).$  
Заменяя неизвестное $(\theta-1)$ на его оценку $\hat\theta-1$, получаем приближённое соотношение  
$\frac{\sqrt{n}\,(\hat\theta-\theta)}{\hat\theta-1}\ \approx\ N(0,1).$  
Следовательно, с вероятностью, близкой к $1-\alpha$,  
$-z < \frac{\sqrt{n}\,(\hat\theta-\theta)}{\hat\theta-1} < z,$  
где $z=z_{1-\alpha/2}$. Преобразуем к неравенству для $\theta$:  
$-z(\hat\theta-1) < \sqrt{n}(\hat\theta-\theta) < z(\hat\theta-1),$  
$\hat\theta-\frac{z(\hat\theta-1)}{\sqrt{n}} < \theta < \hat\theta+\frac{z(\hat\theta-1)}{\sqrt{n}}.$  
Подставляя $\hat\theta = 1+1/\bar L$ ($\bar L=\frac1n\sum\ln X_i$) и $\hat\theta-1=1/\bar L$, получаем рукописную форму:  
$\boxed{1-\frac{z-\sqrt{n}}{\sqrt{n}\,\bar L} \;<\; \theta \;<\; 1+\frac{z+\sqrt{n}}{\sqrt{n}\,\bar L}}.$  
Для $\alpha=0.05$ имеем $z=1.96$.

```python
# Задача 6. Распределение Парето

import numpy as np
import scipy.stats as stats

# Параметры
theta_true = 3.0          # истинное значение θ
n = 100                   # объём выборки
alpha = 0.05              # уровень значимости
beta = 1 - alpha          # доверительная вероятность
z = stats.norm.ppf(1 - alpha/2)   # 1.96
B = 1000                  # число бутстрап-выборок

# Генерация выборки из распределения Парето (θ>1)
# F(x)=1-x^{-(θ-1)} => обратная функция: x = (1-u)^{-1/(θ-1)}
np.random.seed(123)
U = np.random.uniform(0, 1, n)
X = (1 - U) ** (-1/(theta_true - 1))

# Оценка МП
logX = np.log(X)
L_bar = np.mean(logX)           # выборочное среднее логарифмов
theta_hat = 1 + 1 / L_bar
print(f"Истинное θ = {theta_true}")
print(f"Оценка МП: {theta_hat:.4f}")

# ## b) Асимптотический доверительный интервал для медианы

m_hat = 2 ** (1/(theta_hat - 1))
se_m = m_hat * np.log(2) / (theta_hat - 1)
ci_median_as = (m_hat - z * se_m, m_hat + z * se_m)

print("\nАсимптотический 95% ДИ для медианы (п. b):")
print(f"[{ci_median_as[0]:.4f}, {ci_median_as[1]:.4f}]")
print(f"Длина интервала: {ci_median_as[1] - ci_median_as[0]:.4f}")

# c) Асимптотический доверительный интервал для θ

# %%
# 1 - (z - √n)/(√n * L̄) < θ < 1 + (z + √n)/(√n * L̄)
lower = 1 - (z - np.sqrt(n)) / (np.sqrt(n) * L_bar)
upper = 1 + (z + np.sqrt(n)) / (np.sqrt(n) * L_bar)
ci_theta_as = (lower, upper)

print("\nАсимптотический 95% ДИ для θ (по рукописи):")
print(f"[{ci_theta_as[0]:.4f}, {ci_theta_as[1]:.4f}]")
print(f"Длина интервала: {ci_theta_as[1] - ci_theta_as[0]:.4f}")

# Для сравнения приведём стандартную форму:
ci_theta_std = (theta_hat - z*(theta_hat-1)/np.sqrt(n),
                theta_hat + z*(theta_hat-1)/np.sqrt(n))
print(f"\n(Стандартная форма: [{ci_theta_std[0]:.4f}, {ci_theta_std[1]:.4f}])")

# ## e) Бутстраповские доверительные интервалы

# %%
# Непараметрический бутстрап
boot_theta_np = []
for _ in range(B):
    sample = np.random.choice(X, size=n, replace=True)
    L_bar_boot = np.mean(np.log(sample))
    theta_boot = 1 + 1 / L_bar_boot
    boot_theta_np.append(theta_boot)
ci_boot_np = (np.percentile(boot_theta_np, 100*alpha/2),
              np.percentile(boot_theta_np, 100*(1-alpha/2)))

print("\nНепараметрический бутстрап 95% ДИ для θ:")
print(f"[{ci_boot_np[0]:.4f}, {ci_boot_np[1]:.4f}]")
print(f"Длина интервала: {ci_boot_np[1] - ci_boot_np[0]:.4f}")

# Параметрический бутстрап
boot_theta_p = []
for _ in range(B):
    U_boot = np.random.uniform(0, 1, n)
    X_boot = (1 - U_boot) ** (-1/(theta_hat - 1))
    L_bar_boot = np.mean(np.log(X_boot))
    theta_boot = 1 + 1 / L_bar_boot
    boot_theta_p.append(theta_boot)
ci_boot_p = (np.percentile(boot_theta_p, 100*alpha/2),
             np.percentile(boot_theta_p, 100*(1-alpha/2)))

print("\nПараметрический бутстрап 95% ДИ для θ:")
print(f"[{ci_boot_p[0]:.4f}, {ci_boot_p[1]:.4f}]")
print(f"Длина интервала: {ci_boot_p[1] - ci_boot_p[0]:.4f}")

# ## f) Сравнение интервалов
print("\n--- Сравнение интервалов для θ ---")
print(f"Истинное θ = {theta_true}")
print(f"Асимптотический (рукопись): [{ci_theta_as[0]:.4f}, {ci_theta_as[1]:.4f}]  длина = {ci_theta_as[1]-ci_theta_as[0]:.4f}")
print(f"Непарам. бутстрап:         [{ci_boot_np[0]:.4f}, {ci_boot_np[1]:.4f}]  длина = {ci_boot_np[1]-ci_boot_np[0]:.4f}")
print(f"Парам. бутстрап:           [{ci_boot_p[0]:.4f}, {ci_boot_p[1]:.4f}]  длина = {ci_boot_p[1]-ci_boot_p[0]:.4f}")

print("\n--- Накрытие истинного значения ---")
print(f"Асимптотический: {ci_theta_as[0] <= theta_true <= ci_theta_as[1]}")
print(f"Непарам. бутстрап: {ci_boot_np[0] <= theta_true <= ci_boot_np[1]}")
print(f"Парам. бутстрап: {ci_boot_p[0] <= theta_true <= ci_boot_p[1]}")

# Истинная медиана
m_true = 2 ** (1/(theta_true - 1))
print(f"\nИстинная медиана = {m_true:.4f}")
print(f"Асимптотический интервал для медианы: [{ci_median_as[0]:.4f}, {ci_median_as[1]:.4f}]  длина = {ci_median_as[1]-ci_median_as[0]:.4f}")
print(f"Накрытие: {ci_median_as[0] <= m_true <= ci_median_as[1]}")
```
