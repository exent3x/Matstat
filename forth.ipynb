## Задача 4

Случайная величина $\xi$ имеет распределение, заданное смесью непрерывной и дискретной частей:

- на интервалах $(-1,0)$ и $(0,1)$ плотность постоянна и равна $\theta$;
- в точках $0$ и $2$ сосредоточены массы вероятности, каждая равная $\frac12 - \theta$.

Параметр $\theta$ принимает значения в интервале $0 < \theta < \frac12$ (чтобы все вероятности были положительны).

Проверим нормировку: вероятность непрерывной части равна $\int_{-1}^0 \theta\,dx + \int_0^1 \theta\,dx = \theta + \theta = 2\theta$, дискретные точки дают в сумме $2\bigl(\frac12-\theta\bigr)=1-2\theta$, итого $2\theta + 1 - 2\theta = 1$.

### a) Оценки методом моментов и методом максимального правдоподобия

#### Метод моментов
Вычислим первый начальный момент:
$$
\alpha_1 = \mathbb{E}[\xi] = \int_{-1}^0 x\theta\,dx + \int_0^1 x\theta\,dx + 0\cdot\Bigl(\frac12-\theta\Bigr) + 2\cdot\Bigl(\frac12-\theta\Bigr).
$$
Интегралы от нечётной функции по симметричным интервалам равны нулю, поэтому
$$
\alpha_1 = 2\Bigl(\frac12-\theta\Bigr) = 1 - 2\theta.
$$
Приравниваем выборочное среднее $\bar x = \frac1n\sum_{i=1}^n x_i$:
$$
1 - 2\theta = \bar x \quad\Longrightarrow\quad \tilde\theta_{\text{ММ}} = \frac12(1 - \bar x).
$$

#### Метод максимального правдоподобия
Пусть выборка объёма $n$. Обозначим $n_c$ – число наблюдений, попавших в непрерывную часть (интервалы $(-1,0)$ и $(0,1)$). Тогда $n - n_c$ – число наблюдений, равных $0$ или $2$. Функция правдоподобия:
$$
L(\theta) = \theta^{\,n_c} \Bigl(\frac12-\theta\Bigr)^{\,n - n_c}.
$$
Логарифмируем:
$$
\ln L(\theta) = n_c \ln\theta + (n - n_c) \ln\!\Bigl(\frac12-\theta\Bigr).
$$
Производная:
$$
\frac{\partial \ln L}{\partial\theta} = \frac{n_c}{\theta} - \frac{n - n_c}{\frac12-\theta}.
$$
Приравниваем к нулю:
$$
\frac{n_c}{\theta} = \frac{n - n_c}{\frac12-\theta} \quad\Longrightarrow\quad n_c\Bigl(\frac12-\theta\Bigr) = (n - n_c)\theta.
$$
$$
\frac12 n_c - n_c\theta = n\theta - n_c\theta \quad\Longrightarrow\quad \frac12 n_c = n\theta.
$$
Отсюда
$$
\hat\theta_{\text{МП}} = \frac{n_c}{2n}.
$$

### b) Несмещённость и состоятельность

#### Оценка метода моментов
$$
\mathbb{E}[\tilde\theta_{\text{ММ}}] = \frac12\bigl(1 - \mathbb{E}[\bar x]\bigr) = \frac12\bigl(1 - (1-2\theta)\bigr) = \theta,
$$
следовательно, оценка несмещённая.

Найдём дисперсию $\tilde\theta_{\text{ММ}}$:
$$
\mathbb{D}[\tilde\theta_{\text{ММ}}] = \frac14 \mathbb{D}[\bar x] = \frac{1}{4n}\mathbb{D}[\xi].
$$
Вычислим $\mathbb{D}[\xi]$. Второй начальный момент:
$$
\alpha_2 = \int_{-1}^0 x^2\theta\,dx + \int_0^1 x^2\theta\,dx + 0^2\cdot\Bigl(\frac12-\theta\Bigr) + 2^2\cdot\Bigl(\frac12-\theta\Bigr).
$$
Интегралы: $\int_{-1}^0 x^2\,dx = \frac13$, $\int_0^1 x^2\,dx = \frac13$, поэтому
$$
\alpha_2 = \theta\cdot\frac13 + \theta\cdot\frac13 + 4\Bigl(\frac12-\theta\Bigr) = \frac{2\theta}{3} + 2 - 4\theta = 2 - \frac{10}{3}\theta.
$$
Тогда
$$
\mathbb{D}[\xi] = \alpha_2 - \alpha_1^2 = \Bigl(2 - \frac{10}{3}\theta\Bigr) - (1-2\theta)^2 = 2 - \frac{10}{3}\theta - 1 + 4\theta - 4\theta^2 = 1 + \frac{2}{3}\theta - 4\theta^2.
$$
Следовательно,
$$
\mathbb{D}[\tilde\theta_{\text{ММ}}] = \frac{1}{4n}\Bigl(1 + \frac{2}{3}\theta - 4\theta^2\Bigr).
$$
При $n\to\infty$ дисперсия стремится к нулю, значит оценка состоятельна.

#### Оценка максимального правдоподобия
Число непрерывных наблюдений $n_c$ имеет биномиальное распределение $\mathrm{Bin}(n, 2\theta)$, так как вероятность попасть в непрерывную часть равна $2\theta$. Тогда
$$
\mathbb{E}[\hat\theta_{\text{МП}}] = \frac{1}{2n}\mathbb{E}[n_c] = \frac{1}{2n}\cdot 2n\theta = \theta,
$$
оценка несмещённая.
$$
\mathbb{D}[\hat\theta_{\text{МП}}] = \frac{1}{4n^2}\mathbb{D}[n_c] = \frac{1}{4n^2}\cdot 2n\theta(1-2\theta) = \frac{\theta(1-2\theta)}{2n} \to 0,
$$
состоятельна.

### c) Исследование эффективности с помощью неравенства Крамера–Рао

#### Проверка регулярности модели
Рассмотрим функцию правдоподобия одного наблюдения:
$$
p(x;\theta) = 
\begin{cases}
\theta, & x\in(-1,0)\cup(0,1),\\[4pt]
\frac12-\theta, & x=0,\\[4pt]
\frac12-\theta, & x=2.
\end{cases}
$$
Проверим, что можно дифференцировать под знаком интеграла тождества $\int p(x;\theta)\,d\mu(x)=1$. Для этого вычислим
$$
\frac{d}{d\theta}\int p(x;\theta)\,d\mu(x) = \int \frac{\partial p}{\partial\theta}\,d\mu(x).
$$
Производная $\frac{\partial p}{\partial\theta}$:
- на непрерывной части: $\frac{\partial}{\partial\theta}\theta = 1$;
- в точках $0$ и $2$: $\frac{\partial}{\partial\theta}\bigl(\frac12-\theta\bigr) = -1$.
Тогда
$$
\int \frac{\partial p}{\partial\theta}\,d\mu = \int_{-1}^0 1\,dx + \int_0^1 1\,dx + (-1) + (-1) = 1 + 1 - 1 - 1 = 0.
$$
Таким образом, условие дифференцируемости под знаком интеграла выполнено. Функция $\ln p(x;\theta)$ дифференцируема по $\theta$ при $\theta\in(0,\frac12)$ для всех $x$ из носителя, и информация Фишера конечна. Модель регулярна.

#### Информация Фишера
Вычислим производную логарифма:
$$
\frac{\partial \ln p}{\partial\theta} = 
\begin{cases}
\frac{1}{\theta}, & x\in(-1,0)\cup(0,1),\\[6pt]
-\frac{1}{\frac12-\theta}, & x=0,2.
\end{cases}
$$
Тогда информация Фишера для одного наблюдения:
$$
I(\theta) = \mathbb{E}\left[\left(\frac{\partial \ln p}{\partial\theta}\right)^2\right] = \int \left(\frac{\partial \ln p}{\partial\theta}\right)^2 p\,d\mu.
$$
Для непрерывной части: на каждом интервале $\left(\frac{1}{\theta}\right)^2 \cdot \theta = \frac{1}{\theta}$. Интеграл по двум интервалам даёт $2\cdot\frac{1}{\theta} = \frac{2}{\theta}$.
Для дискретных точек: $\left(-\frac{1}{\frac12-\theta}\right)^2 \cdot \left(\frac12-\theta\right) = \frac{1}{\frac12-\theta}$. Две точки дают $\frac{2}{\frac12-\theta}$.
Следовательно,
$$
I(\theta) = \frac{2}{\theta} + \frac{2}{\frac12-\theta} = 2\left(\frac{1}{\theta} + \frac{1}{\frac12-\theta}\right).
$$
Приведём сумму к общему знаменателю:
$$
\frac{1}{\theta} + \frac{1}{\frac12-\theta} = \frac{\frac12-\theta + \theta}{\theta(\frac12-\theta)} = \frac{\frac12}{\theta(\frac12-\theta)} = \frac{1}{2\theta(\frac12-\theta)}.
$$
Таким образом,
$$
I(\theta) = 2 \cdot \frac{1}{2\theta(\frac12-\theta)} = \frac{1}{\theta(\frac12-\theta)} = \frac{2}{\theta(1-2\theta)}.
$$
Оценки регулярны по достаточному условию
#### Неравенство Крамера–Рао
Для выборки объёма $n$ нижняя граница дисперсии несмещённой оценки:
$$
\frac{1}{n I(\theta)} = \frac{\theta(1-2\theta)}{2n}.
$$

#### Сравнение дисперсий
- Для оценки максимального правдоподобия $\hat\theta_{\text{МП}}$:
  $$
  \mathbb{D}[\hat\theta_{\text{МП}}] = \frac{\theta(1-2\theta)}{2n},
  $$
  что в точности совпадает с нижней границей. Следовательно, оценка **эффективна** (достигает границы Крамера–Рао).

- Для оценки метода моментов $\tilde\theta_{\text{ММ}}$:
  $$
  \mathbb{D}[\tilde\theta_{\text{ММ}}] = \frac{1}{4n}\Bigl(1 + \frac{2}{3}\theta - 4\theta^2\Bigr).
  $$
  Вычислим разность с нижней границей:
  $$
  \mathbb{D}[\tilde\theta_{\text{ММ}}] - \frac{\theta(1-2\theta)}{2n} = \frac{1}{4n}\Bigl(1 + \frac{2}{3}\theta - 4\theta^2\Bigr) - \frac{\theta(1-2\theta)}{2n}.
  $$
  Приведём к общему знаменателю $4n$:
  $$
  \frac{1}{4n}\Bigl(1 + \frac{2}{3}\theta - 4\theta^2 - 2\theta(1-2\theta)\Bigr) = \frac{1}{4n}\Bigl(1 + \frac{2}{3}\theta - 4\theta^2 - 2\theta + 4\theta^2\Bigr) = \frac{1}{4n}\Bigl(1 - \frac{4}{3}\theta\Bigr).
  $$
  При $\theta\in(0,\frac12)$ имеем $1 - \frac{4}{3}\theta > 1 - \frac{4}{3}\cdot\frac12 = 1 - \frac{2}{3} = \frac13 > 0$. Следовательно, дисперсия оценки метода моментов строго больше нижней границы, и оценка **неэффективна**.

### Итог
- Оценка метода моментов: $\tilde\theta = \frac12(1-\bar x)$ – несмещённая, состоятельная, но неэффективная.
- Оценка максимального правдоподобия: $\hat\theta = \dfrac{n_c}{2n}$ – несмещённая, состоятельная и эффективная.
