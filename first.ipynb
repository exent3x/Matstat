**Состоятельность исходной оценки:**

Для любого $\varepsilon > 0$

$$
P\bigl(|\hat{\theta}_3 - \theta| > \varepsilon\bigr) = P\bigl(\theta - X_{(n)} > \varepsilon\bigr) = P\bigl(X_{(n)} < \theta - \varepsilon\bigr) = \left(\frac{\theta - \varepsilon}{\theta}\right)^n = \left(1 - \frac{\varepsilon}{\theta}\right)^n.
$$

Поскольку $0 < 1 - \varepsilon/\theta < 1$, то $\left(1 - \frac{\varepsilon}{\theta}\right)^n \to 0$ при $n\to\infty$. Следовательно, $\hat{\theta}_3 \stackrel{p}{\to} \theta$ — оценка состоятельна.

---

**Состоятельность исправленной оценки:**

Используем тот факт, что $X_{(n)} \stackrel{p}{\to} \theta$ и $\frac{n+1}{n} \to 1$. Тогда по теореме о сходимости произведения:

$$
\hat{\theta}_3^* = \frac{n+1}{n}X_{(n)} = X_{(n)} + \frac{1}{n}X_{(n)}.
$$

Для любого $\varepsilon > 0$:

$$
P\bigl(|\hat{\theta}_3^* - \theta| > \varepsilon\bigr) \le P\bigl(|X_{(n)} - \theta| > \varepsilon/2\bigr) + P\bigl(\tfrac{1}{n}X_{(n)} > \varepsilon/2\bigr).
$$

Первое слагаемое стремится к нулю, так как $X_{(n)}$ состоятельна. Второе: $\frac{1}{n}X_{(n)} \le \frac{\theta}{n}$, поэтому для $n > \frac{2\theta}{\varepsilon}$ имеем $\frac{1}{n}X_{(n)} \le \frac{\theta}{n} < \varepsilon/2$, значит вероятность равна нулю. Таким образом, $\hat{\theta}_3^* \stackrel{p}{\to} \theta$, т.е. состоятельна.
