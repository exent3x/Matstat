# $ \text{Т.14} $

$ \text{Пусть } x_n \text{ и } y_m \text{ - независимые случайные выборки} \\ \text{ из нормального распределения с параметрами } a \text{, } \sigma_x^2 = 2 \text{ и } \\ b \text{, } \sigma_y^2 = 1 \text{ соответственно. Используя реализации случайных выборок: } \\ \vec{x} = \begin{pmatrix} -1.11 & -6.10 & 2.42 \end{pmatrix}^T \text{, } \vec{y} = \begin{pmatrix} -2.29 & -2.91 \end{pmatrix}^T \text{,} \\ \text{ проверить гипотезу о равенстве средних против альтернатив } a > b \text{.} $

```python
from rich.console import Console
from rich.panel import Panel
from rich.text import Text

import scipy.stats as stats

import numpy as np
import matplotlib.pyplot as polt
from matplotlib.ticker import AutoMinorLocator

console = Console()
```

```python
alpha = 0.05

n = 3
m = 2

x_n = np.array([-1.11, -6.10, 2.42])
y_m = np.array([-2.29, -2.91])

sigma_x = 2
sigma_y = 1
```

```python
delta = x_n.mean() - y_m.mean()

N = stats.norm(loc=0,
               scale=np.sqrt(sigma_x / n + sigma_y / m))

console.print(Panel(Text(str(delta), style="bold"),
              title="Значение delta"),
              justify="left")
```

```python
q = N.ppf(1 - alpha)
console.print(Panel(Text(str(q), style="bold"),
              title="Квантиль порядка 1 - alpha"),
              justify="left")
```

```python
p_value = N.sf(delta)
console.print(Panel(Text(f"p-value: {p_value}",
                         style="bold"),
                    title="p-value"),
              justify="left")
```

### Задача 14

Дано:  
$X \sim N(a, \sigma_x^2=2)$, выборка $\vec{x}=(-1.11,\; -6.10,\; 2.42)^T$, $n=3$;  
$Y \sim N(b, \sigma_y^2=1)$, выборка $\vec{y}=(-2.29,\; -2.91)^T$, $m=2$.  
Проверить $H_0: a=b$ против $H_1: a>b$.

#### Статистика критерия

При $H_0$ разность выборочных средних  
$$
\Delta = \overline{x} - \overline{y} \sim N\!\left(0,\; \frac{\sigma_x^2}{n}+\frac{\sigma_y^2}{m}\right)=N\!\left(0,\; \frac{2}{3}+\frac{1}{2}\right)=N\!\left(0,\; \frac{7}{6}\right).
$$

#### Критическая область (уровень $\alpha=0.05$)

Односторонняя альтернатива $a>b$:  
$$
G = \left\{ \Delta \ge u_{1-\alpha}\sqrt{\frac{7}{6}} \right\},
$$
где $u_{1-\alpha}$ – квантиль стандартного нормального распределения.  
При $\alpha=0.05$: $u_{0.95}=1.645$,  
$$
\text{критическое значение} = 1.645 \cdot \sqrt{\frac{7}{6}} \approx 1.645 \cdot 1.0801 \approx 1.777.
$$

#### Вычисление наблюдаемого значения

$$
\overline{x} = \frac{-1.11-6.10+2.42}{3} = \frac{-4.79}{3} \approx -1.5967,
$$
$$
\overline{y} = \frac{-2.29-2.91}{2} = \frac{-5.20}{2} = -2.60,
$$
$$
\Delta_{\text{набл}} = \overline{x}-\overline{y} = -1.5967 - (-2.60) = 1.0033.
$$

#### p-value

$$
p = P(\Delta \ge 1.0033 \mid H_0) = 1 - \Phi\!\left(\frac{1.0033}{\sqrt{7/6}}\right) = 1 - \Phi(1.0033/1.0801) = 1 - \Phi(0.929) \approx 1 - 0.8238 = 0.1762.
$$

Так как $p > 0.05$, вывод: нет оснований отвергать H0.

$ \text{Построить график мощности критерия.} $

```python
_, axs = polt.subplots(1, 1, figsize=(10, 6))

axs.set_title("Мощность критерия")
axs.set_xlabel("d")
axs.set_ylabel("W(d)")

d = np.linspace(0, 5, 100)

axs.plot(d, N.sf(-d + q),
         label="Мощность критерия",
         color="green")

axs.xaxis.set_minor_locator(AutoMinorLocator(10))
axs.yaxis.set_minor_locator(AutoMinorLocator(10))

polt.grid(which='minor', linestyle='--', linewidth=0.2)
polt.grid(which='major', linewidth=0.4)

# polt.legend()
polt.show()
```
