# T5




Случайная величина $X$ имеет равномерное распределение на отрезке $[\theta, 2\theta]$, где $\theta > 0$. Плотность распределения:
$$f(x;\theta) = \frac{1}{\theta}, \quad \theta \le x \le 2\theta.$$

---

## a) Оценки методом моментов и методом максимального правдоподобия

**Метод моментов.**  
Первый начальный момент (математическое ожидание):
$$\mathbb{E}[X] = \int_\theta^{2\theta} x \cdot \frac{1}{\theta}\,dx = \frac{1}{\theta} \cdot \frac{(2\theta)^2 - \theta^2}{2} = \frac{1}{\theta} \cdot \frac{4\theta^2 - \theta^2}{2} = \frac{3\theta}{2}.$$
Приравниваем выборочное среднее $\bar{X} = \frac{1}{n}\sum_{i=1}^n X_i$:
$$\bar{X} = \frac{3\theta}{2} \quad \Longrightarrow \quad \hat{\theta}_{\text{ММ}} = \frac{2}{3}\bar{X}.$$

**Метод максимального правдоподобия.**  
Функция правдоподобия для выборки $X_1,\dots,X_n$:
$$L(\theta) = \prod_{i=1}^n \frac{1}{\theta} = \theta^{-n}, \quad \text{при условии } \theta \le \min_{1\le i\le n} X_i \le \max_{1\le i\le n} X_i \le 2\theta.$$
Условия существования: $\theta \le X_{(1)}$ и $X_{(n)} \le 2\theta$, где $X_{(1)} = \min X_i$, $X_{(n)} = \max X_i$. Таким образом, $\theta$ должно удовлетворять:
$$\theta \le X_{(1)}, \quad \theta \ge \frac{X_{(n)}}{2}.$$
Функция $L(\theta) = \theta^{-n}$ убывает с ростом $\theta$, поэтому максимум достигается при наименьшем возможном $\theta$, но с учётом ограничения $\theta \ge X_{(n)}/2$. Следовательно,
$$\hat{\theta}_{\text{МП}} = \frac{X_{(n)}}{2}.$$
---

## b) Несмещённость и состоятельность. Исправление оценок

**Оценка метода моментов.**  
$$\mathbb{E}[\hat{\theta}_{\text{ММ}}] = \frac{2}{3}\mathbb{E}[\bar{X}] = \frac{2}{3} \cdot \frac{3\theta}{2} = \theta,$$
следовательно, оценка несмещённая.

Дисперсия:
$$\mathbb{D}[\hat{\theta}_{\text{ММ}}] = \left(\frac{2}{3}\right)^2 \mathbb{D}[\bar{X}] = \frac{4}{9} \cdot \frac{1}{n} \mathbb{D}[X].$$
Вычислим $\mathbb{D}[X]$. Второй начальный момент:
$$\mathbb{E}[X^2] = \int_\theta^{2\theta} x^2 \cdot \frac{1}{\theta}\,dx = \frac{1}{\theta} \cdot \frac{(2\theta)^3 - \theta^3}{3} = \frac{1}{\theta} \cdot \frac{8\theta^3 - \theta^3}{3} = \frac{7\theta^2}{3}.$$
Тогда
$$\mathbb{D}[X] = \mathbb{E}[X^2] - (\mathbb{E}[X])^2 = \frac{7\theta^2}{3} - \left(\frac{3\theta}{2}\right)^2 = \frac{7\theta^2}{3} - \frac{9\theta^2}{4} = \theta^2\left(\frac{28}{12} - \frac{27}{12}\right) = \frac{\theta^2}{12}.$$
Таким образом,
$$\mathbb{D}[\hat{\theta}_{\text{ММ}}] = \frac{4}{9} \cdot \frac{1}{n} \cdot \frac{\theta^2}{12} = \frac{\theta^2}{27n}.$$
При $n\to\infty$ дисперсия стремится к нулю, значит оценка состоятельна.

**Оценка максимального правдоподобия.**  
$$F_Y(y) = \left(\frac{y-\theta}{\theta}\right)^n, \quad \theta \le y \le 2\theta.$$
Плотность:
$$f_Y(y) = n \left(\frac{y-\theta}{\theta}\right)^{n-1} \cdot \frac{1}{\theta} = \frac{n}{\theta^n} (y-\theta)^{n-1}, \quad \theta \le y \le 2\theta.$$
Математическое ожидание:
$$\mathbb{E}[Y] = \int_\theta^{2\theta} y \cdot \frac{n}{\theta^n} (y-\theta)^{n-1} dy.$$
Сделаем замену $t = y-\theta$, тогда $y = t+\theta$, $dy = dt$, $t \in [0, \theta]$:
$$\mathbb{E}[Y] = \frac{n}{\theta^n} \int_0^\theta (t+\theta) t^{n-1} dt = \frac{n}{\theta^n} \left( \int_0^\theta t^n dt + \theta \int_0^\theta t^{n-1} dt \right) = \frac{n}{\theta^n} \left( \frac{\theta^{n+1}}{n+1} + \theta \cdot \frac{\theta^n}{n} \right) = \frac{n}{\theta^n} \left( \frac{\theta^{n+1}}{n+1} + \frac{\theta^{n+1}}{n} \right) = n\theta \left( \frac{1}{n+1} + \frac{1}{n} \right) = n\theta \cdot \frac{n + n+1}{n(n+1)} = \theta \cdot \frac{2n+1}{n+1}.$$
Следовательно,
$$\mathbb{E}[\hat{\theta}_{\text{МП}}] = \mathbb{E}\left[\frac{Y}{2}\right] = \frac{\theta}{2} \cdot \frac{2n+1}{n+1} = \theta \cdot \frac{2n+1}{2(n+1)} \neq \theta,$$
оценка смещённая.
**Исправление оценки МП.**  
 исправленная оценка:
$$\hat{\theta}_{\text{МП}}^* = \frac{n+1}{2n+1} \cdot Y = \frac{n+1}{2n+1} \max X_i.$$
Проверим несмещённость:
$$\mathbb{E}[\hat{\theta}_{\text{МП}}^*] = \frac{n+1}{2n+1} \cdot \theta \cdot \frac{2n+1}{n+1} = \theta.$$

Найдём дисперсию исправленной оценки. Сначала вычислим $\mathbb{E}[Y^2]$:
$$\mathbb{E}[Y^2] = \int_\theta^{2\theta} y^2 \cdot \frac{n}{\theta^n} (y-\theta)^{n-1} dy.$$
Та же замена $t = y-\theta$:
$$\mathbb{E}[Y^2] = \frac{n}{\theta^n} \int_0^\theta (t+\theta)^2 t^{n-1} dt = \frac{n}{\theta^n} \int_0^\theta (t^2 + 2\theta t + \theta^2) t^{n-1} dt = \frac{n}{\theta^n} \left( \int_0^\theta t^{n+1} dt + 2\theta \int_0^\theta t^n dt + \theta^2 \int_0^\theta t^{n-1} dt \right).$$
Вычисляем интегралы:
$$\int_0^\theta t^{n+1} dt = \frac{\theta^{n+2}}{n+2}, \quad \int_0^\theta t^n dt = \frac{\theta^{n+1}}{n+1}, \quad \int_0^\theta t^{n-1} dt = \frac{\theta^n}{n}.$$
Подставляем:
$$\mathbb{E}[Y^2] = \frac{n}{\theta^n} \left( \frac{\theta^{n+2}}{n+2} + 2\theta \cdot \frac{\theta^{n+1}}{n+1} + \theta^2 \cdot \frac{\theta^n}{n} \right) = n \left( \frac{\theta^2}{n+2} + \frac{2\theta^2}{n+1} + \frac{\theta^2}{n} \right) = \theta^2 \left( \frac{n}{n+2} + \frac{2n}{n+1} + 1 \right).$$
Приведём к общему знаменателю $\mathbb{D}[Y] = \mathbb{E}[Y^2] - (\mathbb{E}[Y])^2$. Имеем:
$$\mathbb{E}[Y] = \theta \cdot \frac{2n+1}{n+1}, \quad (\mathbb{E}[Y])^2 = \theta^2 \cdot \frac{(2n+1)^2}{(n+1)^2}.$$
Тогда
$$\mathbb{D}[Y] = \theta^2 \left( \frac{n}{n+2} + \frac{2n}{n+1} + 1 - \frac{(2n+1)^2}{(n+1)^2} \right).$$
 Обозначим $A = \frac{n}{n+2} + \frac{2n}{n+1} + 1$. Приведём первые два слагаемых к общему знаменателю $(n+1)(n+2)$:
$$\frac{n}{n+2} = \frac{n(n+1)}{(n+1)(n+2)}, \quad \frac{2n}{n+1} = \frac{2n(n+2)}{(n+1)(n+2)}.$$
Тогда
$$A = \frac{n(n+1) + 2n(n+2)}{(n+1)(n+2)} + 1 = \frac{n^2+n + 2n^2+4n}{(n+1)(n+2)} + 1 = \frac{3n^2+5n}{(n+1)(n+2)} + 1 = \frac{3n^2+5n + (n+1)(n+2)}{(n+1)(n+2)}.$$
Вычислим $(n+1)(n+2) = n^2+3n+2$. Тогда числитель:
$$3n^2+5n + n^2+3n+2 = 4n^2+8n+2 = 2(2n^2+4n+1).$$
Таким образом,
$$A = \frac{2(2n^2+4n+1)}{(n+1)(n+2)}.$$
Теперь вычтем $\frac{(2n+1)^2}{(n+1)^2}$. Приведём к общему знаменателю $(n+1)^2(n+2)$:
$$\mathbb{D}[Y]/\theta^2 = \frac{2(2n^2+4n+1)}{(n+1)(n+2)} - \frac{(2n+1)^2}{(n+1)^2} = \frac{2(2n^2+4n+1)(n+1) - (2n+1)^2(n+2)}{(n+1)^2(n+2)}.$$
Раскроем скобки:
$$2(2n^2+4n+1)(n+1) = 2(2n^3+2n^2+4n^2+4n + n+1) = 2(2n^3+6n^2+5n+1) = 4n^3+12n^2+10n+2.$$
$$(2n+1)^2(n+2) = (4n^2+4n+1)(n+2) = 4n^3+8n^2+4n^2+8n + n+2 = 4n^3+12n^2+9n+2.$$
Вычитаем:
$$(4n^3+12n^2+10n+2) - (4n^3+12n^2+9n+2) = n.$$
Таким образом,
$$\mathbb{D}[Y] = \theta^2 \cdot \frac{n}{(n+1)^2(n+2)}.$$
Следовательно, дисперсия исправленной оценки:
$$\mathbb{D}[\hat{\theta}_{\text{МП}}^*] = \left(\frac{n+1}{2n+1}\right)^2 \mathbb{D}[Y] = \left(\frac{n+1}{2n+1}\right)^2 \cdot \frac{n\theta^2}{(n+1)^2(n+2)} = \frac{n\theta^2}{(2n+1)^2 (n+2)}.$$
При $n\to\infty$ дисперсия стремится к нулю, значит оценка состоятельна.

---

## c) Сравнение эффективности исправленных оценок

Имеем две несмещённые оценки:
$$\hat{\theta}_{\text{ММ}} = \frac{2}{3}\bar{X}, \quad \mathbb{D}[\hat{\theta}_{\text{ММ}}] = \frac{\theta^2}{27n};$$
$$\hat{\theta}_{\text{МП}}^* = \frac{n+1}{2n+1} \max X_i, \quad \mathbb{D}[\hat{\theta}_{\text{МП}}^*] = \frac{n\theta^2}{(2n+1)^2 (n+2)}.$$

Сравним их асимптотически при больших $n$. Для МП-оценки:
$$\mathbb{D}[\hat{\theta}_{\text{МП}}^*] \sim \frac{n\theta^2}{(2n)^2 \cdot n} = \frac{n\theta^2}{4n^3} = \frac{\theta^2}{4n^2},$$
Для равномерного распределения носитель $[\theta, 2\theta]$ зависит от параметра, поэтому условия регулярности не выполнены: область определения зависит от $\theta$, и нельзя менять порядок интегрирования и дифференцирования. Неравенство Крамера–Рао неприменимо.Сравнение дисперсий показывает преимущество МП-оценки.

## d) Точный доверительный интервал для параметра $\theta$

### 1. Выбор центральной статистики
Рассмотрим статистику $Y = \max(X_1, \dots, X_n)$
$$F_X(x) = \frac{x - \theta}{\theta}, \quad \theta \le x \le 2\theta.$$
Тогда для максимума:
$$F_Y(y) = P(Y \le y) = \left( \frac{y - \theta}{\theta} \right)^n, \quad \theta \le y \le 2\theta.$$
Плотность:
$$f_Y(y) = n \left( \frac{y - \theta}{\theta} \right)^{n-1} \cdot \frac{1}{\theta} = \frac{n}{\theta^n} (y - \theta)^{n-1}.$$

### 2. Переход к величине с известным распределением
Сделаем преобразование:
$$Z = \frac{Y - \theta}{\theta}.$$
Тогда $Z$ принимает значения от $0$ до $1$.
$$F_Z(z) = P\left( \frac{Y - \theta}{\theta} \le z \right) = P(Y \le \theta + \theta z) = \left( \frac{\theta + \theta z - \theta}{\theta} \right)^n = z^n, \quad 0 \le z \le 1.$$
Это функция распределения бета-распределения первого рода с параметрами $(n, 1)$, т.е. $Z \sim \mathrm{Beta}(n,1)$. Её плотность $f_Z(z) = n z^{n-1}$, $0 \le z \le 1$.

### 3. Построение доверительного интервала для $Z$
Для заданной доверительной вероятности $\beta$ выберем такие числа $z_1$ и $z_2$, чтобы
$$P(z_1 < Z < z_2) = \beta.$$
$$P(Z \le z_1) = \frac{1-\beta}{2}, \quad P(Z \le z_2) = \frac{1+\beta}{2}.$$
Используя функцию распределения $F_Z(z) = z^n$, получаем:
$$z_1^n = \frac{1-\beta}{2} \quad \Rightarrow \quad z_1 = \left( \frac{1-\beta}{2} \right)^{1/n},$$
$$z_2^n = \frac{1+\beta}{2} \quad \Rightarrow \quad z_2 = \left( \frac{1+\beta}{2} \right)^{1/n}.$$
Таким образом,
$$P\left( \left( \frac{1-\beta}{2} \right)^{1/n} < Z < \left( \frac{1+\beta}{2} \right)^{1/n} \right) = \beta.$$

### 4. Переход к неравенству для $\theta$
$$\left( \frac{1-\beta}{2} \right)^{1/n} < \frac{Y}{\theta} - 1 < \left( \frac{1+\beta}{2} \right)^{1/n}.$$
Прибавляем 1 ко всем частям:
$$1 + \left( \frac{1-\beta}{2} \right)^{1/n} < \frac{Y}{\theta} < 1 + \left( \frac{1+\beta}{2} \right)^{1/n}.$$

$$\frac{1}{1 + \left( \frac{1+\beta}{2} \right)^{1/n}} < \frac{\theta}{Y} < \frac{1}{1 + \left( \frac{1-\beta}{2} \right)^{1/n}}.$$
Умножаем на $Y$:
$$\frac{Y}{1 + \left( \frac{1+\beta}{2} \right)^{1/n}} < \theta < \frac{Y}{1 + \left( \frac{1-\beta}{2} \right)^{1/n}}.$$

### 5. Окончательная формула точного доверительного интервала
$$\boxed{ \theta \in \left( \frac{\max X_i}{1 + \left( \frac{1+\beta}{2} \right)^{1/n}},\; \frac{\max X_i}{1 + \left( \frac{1-\beta}{2} \right)^{1/n}} \right) }.$$

## e) Асимптотический доверительный интервал для $\theta$

### 1. Оценка метода моментов
$$\tilde{\theta} = \frac{2}{3} \bar{X}, \quad \text{где } \bar{X} = \frac{1}{n}\sum_{i=1}^n X_i.$$

### 2. Применение центральной предельной теоремы
По ЦПТ для выборочного среднего:
$$\sqrt{n}(\bar{X} - \mathbb{E}[X]) \xrightarrow{d} N(0, \mathbb{D}[X]),$$
где $\mathbb{E}[X] = \frac{3\theta}{2}$, $\mathbb{D}[X] = \frac{\theta^2}{12}$. Тогда:
$$\sqrt{n}\left( \bar{X} - \frac{3\theta}{2} \right) \xrightarrow{d} N\left(0, \frac{\theta^2}{12}\right).$$

Выразим отклонение оценки $\tilde{\theta}$ от истинного $\theta$:
$$\tilde{\theta} - \theta = \frac{2}{3}\bar{X} - \theta = \frac{2}{3}\left( \bar{X} - \frac{3\theta}{2} \right).$$
Следовательно,
$$\sqrt{n}(\tilde{\theta} - \theta) = \frac{2}{3} \sqrt{n}\left( \bar{X} - \frac{3\theta}{2} \right) \xrightarrow{d} \frac{2}{3} N\left(0, \frac{\theta^2}{12}\right) = N\left(0, \frac{4}{9} \cdot \frac{\theta^2}{12}\right) = N\left(0, \frac{\theta^2}{27}\right).$$

### 3. Построение интервала
$$\frac{3n(\tilde{\theta} - \theta)}{2\sqrt{(\tilde\alpha_2-\tilde\alpha_1^2)}}.$$
Она асимптотически распределена как $N(0,1)$. Для доверительной вероятности $\beta = 0.95$ квантили $\pm 1.96$:
$$P\left( -1.96 < \frac{3n(\tilde{\theta} - \theta)}{2\sqrt{S^2(n-1)}} < 1.96 \right) = \beta.$$
Отсюда выражаем $\theta$:
$$-1.96 < \frac{3n(\tilde{\theta} - \theta)}{2\sqrt{S^2(n-1)}} < 1.96$$
$$\Rightarrow \quad \tilde{\theta} - \frac{1.96 \cdot 2\sqrt{S^2(n-1)}}{3n} < \theta < \tilde{\theta} + \frac{1.96 \cdot 2\sqrt{S^2(n-1)}}{3n}.$$

$$\boxed{ \frac{2}{3}\bar{X} - \frac{3.92 \sqrt{S^2(n-1)}}{3n} < \theta < \frac{2}{3}\bar{X} + \frac{3.92 \sqrt{S^2(n-1)}}{3n} }.$$


```python
import numpy as np
import scipy.stats as stats

theta_true = 5.0
n = 100
alpha = 0.05
# alpha=0.05 аналогично b=0.95

B = 1000
np.random.seed(123)
X = np.random.uniform(theta_true, 2*theta_true, n)

theta_mm = (2/3) * np.mean(X)
theta_ml_corr = ((n+1)/(2*n+1)) * np.max(X)

print(f"Истинное theta = {theta_true}")
print(f"Оценка ММ: {theta_mm:.4f}")
print(f"Оценка МП (исправленная): {theta_ml_corr:.4f}")

Y = np.max(X)
gamma1 = (alpha/2)**(1/n)
gamma2 = (1 - alpha/2)**(1/n)
ci_exact = (Y/(1+gamma2), Y/(1+gamma1))
print(f"\nТочный 95% ДИ: [{ci_exact[0]:.4f}, {ci_exact[1]:.4f}]")

# Асимптотический доверительный интервал 
z = stats.norm.ppf(1 - alpha/2)
se_as = theta_mm / np.sqrt(27*n)
ci_as = (theta_mm - z * se_as, theta_mm + z * se_as)
print(f"Асимптотический 95% ДИ (ММ): [{ci_as[0]:.4f}, {ci_as[1]:.4f}]")

# Бутстраповский доверительный интервал 
bootstrap_estimates = []
for _ in range(B):
    sample = np.random.choice(X, size=n, replace=True)
    boot_theta = ((n+1)/(2*n+1)) * np.max(sample)
    bootstrap_estimates.append(boot_theta)

ci_boot_perc = (np.percentile(bootstrap_estimates, 100*alpha/2),
                np.percentile(bootstrap_estimates, 100*(1-alpha/2)))
print(f"Бутстрап 95% ДИ: [{ci_boot_perc[0]:.4f}, {ci_boot_perc[1]:.4f}]")

print("\nСравнение интервалов:")
print(f"Точный: длина = {ci_exact[1]-ci_exact[0]:.4f}")
print(f"Асимптотический: длина = {ci_as[1]-ci_as[0]:.4f}")
print(f"Бутстрап: длина = {ci_boot_perc[1]-ci_boot_perc[0]:.4f}")

print("\nНакрытие истинного значения theta=5:")
print(f"Точный: {ci_exact[0] <= theta_true <= ci_exact[1]}")
print(f"Асимптотический: {ci_as[0] <= theta_true <= ci_as[1]}")
print(f"Бутстрап: {ci_boot_perc[0] <= theta_true <= ci_boot_perc[1]}")
```
