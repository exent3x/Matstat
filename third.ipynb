

## Задача 3

Пусть \(X\) имеет экспоненциальное распределение с плотностью  
$f(x;\theta)=\frac{1}{\theta}e^{-x/\theta},\quad x\ge 0,\  f(x,\theta)=0,\quad x<0\ \theta>0.$ 
Объём выборки \(n=3\).
### Оценки  
$ \tilde\theta_1=\bar x$ и $\tilde\theta_3=x_{(2)}.$

### a) Проверка на несмещённость

1. $\tilde\theta_1=\bar x$:  
   $\mathbb{E}[\bar x]=\mathbb{E}[\frac{1}{n}\sum_{i=1}^{n} x_i]=\frac{1}{n}\mathbb{E}[x]=n*\theta*\frac{1}{n}=\theta\;\Rightarrow\; \text{несмещённая}.$

2. $\tilde\theta_3=x_{(2)}$:  
   $f_{(2)}(x)=\frac{3!}{1!\,1!}\,F(x)(1-F(x))f(x)$ = $ 6*(1-e^{\frac{-x}{\theta}})*e^{\frac{-x}{\theta}}\cdot\frac{1}{\theta}e^{\frac{-x}{\theta}}$
   = $\frac{6}{\theta}\left(1-e^{-x/\theta}\right)e^{-2x/\theta},\quad x\ge0 $.
   #### Вычислим математическое ожидание:
   $\mathbb{E}[X_{(2)}]=\int_0^\infty x f_{(2)}(x)\,dx
   =6\theta\int_0^\infty t(1-e^{-t})e^{-2t}\,dt
   =6\theta\left(\frac{1}{4}-\frac{1}{9}\right)=\frac{5}{6}\theta.$
   Оценка смещённая. 
   Исправленная оценка: $\hat\theta_3=\frac{6}{5}x_{(2)}$, тогда $\mathbb{E}[\hat\theta_3]=\theta$.

### b) Сравнение эффективности исправленных оценок

Дисперсии:

- $\hat\theta_1=\bar x$:  
  $\mathbb{D}[\hat\theta_1]=\frac{\theta^2}{3}$. по предыдущим номерам

- Для $\hat\theta_3=\frac{6}{5}x_{(2)}$ найдём $\mathbb{D}[X_{(2)}]$:  
  $\mathbb{E}[X_{(2)}^2]=\int_0^\infty x^2 f_{(2)}(x)\,dx
  =6\theta^2\int_0^\infty t^2(1-e^{-t})e^{-2t}\,dt
  =6\theta^2\left(\frac{1}{4}-\frac{2}{27}\right)=\frac{19}{18}\theta^2$.
  Тогда  
  $\mathbb{D}[X_{(2)}]=\frac{19}{18}\theta^2-\left(\frac{5}{6}\theta\right)^2=\frac{13}{36}\theta^2$.
  Следовательно,
  $\mathbb{D}[\hat\theta_3]=\left(\frac{6}{5}\right)^2\cdot\frac{13}{36}\theta^2=\frac{13}{25}\theta^2$.

  Сравниваем: $\frac{13}{25}=0.52>\frac{1}{3}\approx0.333$. Оценка $\hat\theta_1$ эффективнее.

### c) Исследование эффективности с помощью неравенства Крамера–Рао

**Проверка условий регулярности для экспоненциального семейства:**

- $(0,\infty)$ не зависит от параметра $(\theta)$.
- $\ln f(x;\theta)=-\ln\theta-\frac{x}{\theta}$ непрерывно дифференцируема по $(\theta)$ при $\theta>0$ для всех $x$.

### Проверим свойства интегралов

Для экспоненциального распределения с плотностью
$
f(x;\theta)=\frac{1}{\theta}e^{-x/\theta},\quad x\ge 0,\ \theta>0,
$
выполняется
$
\int_0^\infty f(x;\theta)\,dx = 1.
$

$ \frac{d}{d\theta}\int_0^\infty \frac{1}{\theta}e^{-x/\theta}\,dx
= \int_0^\infty \frac{\partial}{\partial\theta}\left(\frac{1}{\theta}e^{-x/\theta}\right)dx = 0.$

#### Вычислим производную подынтегрального выражения:
$\frac{\partial}{\partial\theta}\left(\frac{1}{\theta}e^{-x/\theta}\right)
= -\frac{1}{\theta^2}e^{-x/\theta} + \frac{1}{\theta}e^{-x/\theta}\cdot\frac{x}{\theta^2}
= e^{-x/\theta}\left(-\frac{1}{\theta^2}+\frac{x}{\theta^3}\right)
= \frac{e^{-x/\theta}}{\theta^3}(x-\theta)$.

#### Интеграл от этой производной:
$\int_0^\infty \frac{e^{-x/\theta}}{\theta^3}(x-\theta)\,dx
= \frac{1}{\theta^2}\int_0^\infty e^{-t}(t-1)\,dt \quad (t=x/\theta)
= \frac{1}{\theta^2}\bigl(1-1\bigr)=0.$

#### Таким образом:
$\int_0^\infty \frac{\partial}{\partial\theta}\left(\frac{1}{\theta}e^{-x/\theta}\right)dx = 0$
---


- $\frac{\partial}{\partial\theta}\ln f(x;\theta)=-\frac{1}{\theta}+\frac{x}{\theta^2}$.
- Дисперсии обоих оценок ограничены
- Исправленные оценки несмещены
- Существует $\mathbb{E}_\theta\left[\left(\frac{\partial}{\partial\theta}\ln f(X;\theta)\right)^2\right]$:
  $\mathbb{E}\left[\left(\frac{\partial\ln f}{\partial\theta}\right)^2\right]=\mathbb{E}\left[\left(\frac{X-\theta}{\theta^2}\right)^2\right]=\frac{\mathbb{D}[X]}{\theta^4}=\frac{\theta^2}{\theta^4}=\frac{1}{\theta^2}.$
- Модель регулярна по сказанному выше
- По достаточному условию оценки регулярны

**Информация Фишера для одного наблюдения:**  
$I(\theta)=\frac{1}{\theta^2}$.  
Для выборки объёма \(n=3\) нижняя граница дисперсии несмещённой оценки:  
$\frac{1}{nI(\theta)}=\frac{\theta^2}{3}$.

- Оценка $\hat\theta_1=\bar x$: $\mathbb{D}[\hat\theta_1]=\frac{\theta^2}{3}$ – достигает границы $\Rightarrow$ эффективна.
- Оценка $\hat\theta_3=\frac{6}{5}x_{(2)}$: $\mathbb{D}[\hat\theta_3]=\frac{13}{25}\theta^2>\frac{\theta^2}{3}$ – неэффективна.
