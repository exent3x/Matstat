

## Задача 8

Проверяется гипотеза $H_0$: номер партии и размер детали независимы. Данные представлены в таблице сопряжённости $2\times 3$:

| Партия | Заниженный | Точный | Завышенный | Сумма |
|--------|------------|--------|------------|-------|
| 1      | 25         | 50     | 25         | 100   |
| 2      | 52         | 41     | 7          | 100   |
| Сумма  | 77         | 91     | 32         | 200   |

### 1. Расчёт ожидаемых частот при условии независимости

Ожидаемая частота $E_{ij} = \frac{\text{итог строки}_i \cdot \text{итог столбца}_j}{n}$:
- $E_{11} = \frac{100\cdot77}{200} = 38{,}5$,
- $E_{12} = \frac{100\cdot91}{200} = 45{,}5$,
- $E_{13} = \frac{100\cdot32}{200} = 16$,
- $E_{21} = \frac{100\cdot77}{200} = 38{,}5$,
- $E_{22} = \frac{100\cdot91}{200} = 45{,}5$,
- $E_{23} = \frac{100\cdot32}{200} = 16$.

### 2. Статистика $\chi^2$

$$
\chi^2_{\text{набл}} = \sum_{i=1}^{2}\sum_{j=1}^{3} \frac{(n_{ij} - E_{ij})^2}{E_{ij}}.
$$
Вычислим слагаемые:
- $(25-38{,}5)^2 / 38{,}5 = 182{,}25 / 38{,}5 \approx 4{,}734$,
- $(50-45{,}5)^2 / 45{,}5 = 20{,}25 / 45{,}5 \approx 0{,}445$,
- $(25-16)^2 / 16 = 81 / 16 = 5{,}0625$,
- $(52-38{,}5)^2 / 38{,}5 = 182{,}25 / 38{,}5 \approx 4{,}734$,
- $(41-45{,}5)^2 / 45{,}5 = 20{,}25 / 45{,}5 \approx 0{,}445$,
- $(7-16)^2 / 16 = 81 / 16 = 5{,}0625$.

Сумма: $\chi^2_{\text{набл}} \approx 4{,}734 + 0{,}445 + 5{,}0625 + 4{,}734 + 0{,}445 + 5{,}0625 = 20{,}483$.

### 3. Число степеней свободы

Для таблицы $n\times m$: $\text{df} = (n-1)(m-1) = (2-1)(3-1) = 2$.

### 4. Критическое значение и вывод

Критическое значение $\chi^2_{0{,}05}(2) \approx 5{,}991$.  
$\chi^2_{\text{набл}} = 20{,}48 > 5{,}991$  

Статистика $\chi^2_{\text{набл}} = 20{,}48$ имеет распределение $\chi^2$ с $2$ степенями свободы.  
Плотность $f_{\chi^2(2)}(t) = \frac12 e^{-t/2}$, поэтому

$$
\text{P-value} = \int_{20{,}48}^{+\infty} \frac12 e^{-t/2}\,dt.
$$

Вычислим интеграл:
$$
\int_{a}^{+\infty} \frac12 e^{-t/2}\,dt = \left[-e^{-t/2}\right]_{a}^{+\infty} = e^{-a/2}.
$$

Подставляем $a = 20{,}48$:
$$
\text{P-value} = e^{-20{,}48 / 2} = e^{-10{,}24}.
$$

Численное значение:
$$
e^{-10} \approx 4{,}53999 \cdot 10^{-5}, \quad e^{-0{,}24} \approx 0{,}78663,
$$
$$
e^{-10{,}24} = e^{-10} \cdot e^{-0{,}24} \approx 4{,}53999 \cdot 10^{-5} \cdot 0{,}78663 \approx 3{,}571 \cdot 10^{-5}.
$$

Таким образом, P‑value $\approx 3{,}57 \cdot 10^{-5}$, что значительно меньше $\alpha = 0{,}05$.
$H_0$ отвергается

**Вывод:** на уровне значимости 0,05 можно считать, что номер партии и размер детали зависимы. Доля деталей с завышенным размером во второй партии существенно меньше, чем в первой.
