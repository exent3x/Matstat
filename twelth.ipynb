### Решение задачи 12

$H_0$ (гипотеза игрока):  
$p_1 = p_2 = \dfrac14,\; p_3 = \dfrac16,\; p_4 = \dfrac13$ (вероятности граней 1,2,3,4).  

$H_1$ (честная кость):  
$q_i = \dfrac14$ для $i=1,2,3,4$.  

Объём выборки $n=2$, уровень значимости $\alpha = 0.2$.

Наиболее мощный критерий основан на отношении правдоподобия  
$\Lambda = \dfrac{L_1}{L_0} = \dfrac{(1/4)^2}{p_{x_1}p_{x_2}} = \dfrac{1}{16\,p_{x_1}p_{x_2}}$.  
Критическая область $G$ состоит из исходов с наибольшими значениями $\Lambda$.

Вычислим $\Lambda$ для всех упорядоченных пар $(x_1,x_2)$:

| Исход | $p_{x_1}p_{x_2}$ | $\Lambda$ |
|-------|------------------|------------|
| (3,3) | $1/36$ | $36/16 = 2.25$ |
| (1,3),(3,1),(2,3),(3,2) | $1/24$ | $24/16 = 1.5$ |
| (3,4),(4,3) | $1/18$ | $18/16 = 1.125$ |
| (1,1),(1,2),(2,1),(2,2) | $1/16$ | $1$ |
| (1,4),(4,1),(2,4),(4,2) | $1/12$ | $12/16 = 0.75$ |
| (4,4) | $1/9$ | $9/16 = 0.5625$ |

Упорядочим по убыванию $\Lambda$ и накопим вероятности при $H_0$:

- $\Lambda = 2.25$: $(3,3)$, $P_0 = 1/36 \approx 0.0278$  
- $\Lambda = 1.5$: $(1,3),(3,1),(2,3),(3,2)$, сумма $P_0 = 4/24 = 1/6 \approx 0.1667$  

Общая вероятность $1/36 + 1/6 = 7/36 \approx 0.1944 \le 0.2$.  
Добавление следующей группы ($\Lambda=1.125$) дало бы $P_0 = 7/36 + 1/9 = 11/36 \approx 0.3056 > 0.2$.


$$
G = \bigl\{(3,3),\;(1,3),\;(3,1),\;(2,3),\;(3,2)\bigr\}.
$$

Ошибка первого рода $\alpha_1 = P_0(G) = \dfrac{7}{36}$.

При $H_1$ все 16 упорядоченных исходов равновероятны с вероятностью $1/16$.  
Мощность критерия (вероятность отвергнуть $H_0$ при $H_1$):

$$
W = P_1(G) = \frac{5}{16}.
$$

**Ответ:** критическая область $G = \{(3,3), (1,3), (3,1), (2,3), (3,2)\}$,  
$\alpha_1 = \dfrac{7}{36}$, мощность $W = \dfrac{5}{16}$.
