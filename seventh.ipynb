## Задача 7

Проверяется гипотеза $H_0$: число смертей в одном корпусе за год подчиняется распределению Пуассона с неизвестным параметром $\lambda$. Данные наблюдений:

| $i$ – число смертей | 0 | 1 | 2 | 3 | 4 |
|---------------------|---|---|---|---|---|
| Число случаев $m_i$ |109|65|22|3|1|

Объём выборки $n = 109+65+22+3+1 = 200$.

### 1. Оценка параметра $\lambda$

Методом максимального правдоподобия получаем оценку  
$$
\hat{\lambda} = \frac{1}{n}\sum_{i} i \cdot m_i = \frac{0\cdot109 + 1\cdot65 + 2\cdot22 + 3\cdot3 + 4\cdot1}{200} = \frac{122}{200} = 0{,}61.
$$

### 2. Группирование данных

Для применения критерия $\chi^2$ необходимо, чтобы ожидаемые частоты были не менее 5. Поэтому объединим группы $i\ge 2$ в одну.  
Наблюдаемые частоты после группирования:
- $m_0 = 109$,
- $m_1 = 65$,
- $m_{\ge2} = 22+3+1 = 26$.

### 3. Ожидаемые частоты при $H_0$

Вероятности по закону Пуассона с $\hat{\lambda}=0{,}61$:
$$
p_0 = e^{-0{,}61} \approx 0{,}54335, \quad
p_1 = 0{,}61\cdot e^{-0{,}61} \approx 0{,}33144, \quad
p_{\ge2} = 1 - p_0 - p_1 \approx 0{,}12521.
$$
Ожидаемые частоты:
$$
n p_0 \approx 200 \cdot 0{,}54335 = 108{,}67, \quad
n p_1 \approx 200 \cdot 0{,}33144 = 66{,}29, \quad
n p_{\ge2} \approx 200 \cdot 0{,}12521 = 25{,}04.
$$
Все ожидаемые частоты больше 5, условие применимости критерия выполнено.

### 4. Расчёт статистики $\chi^2$

$$
\chi^2_{\text{набл}} = \sum_{j} \frac{(m_j - n p_j)^2}{n p_j} = 
\frac{(109 - 108{,}67)^2}{108{,}67} + \frac{(65 - 66{,}29)^2}{66{,}29} + \frac{(26 - 25{,}04)^2}{25{,}04} \approx 0{,}0010 + 0{,}0251 + 0{,}0368 = 0{,}0629.
$$

### 5. Число степеней свободы

После группирования имеем $k=3$ группы. Оценён один параметр ($\lambda$), поэтому  
$$
\text{df} = k - 1 - 1 = 1.
$$

### 6. Критическое значение и вывод

Статистика $\chi^2_{\text{набл}} = 0{,}0629$ имеет распределение $\chi^2$ с $1$ степенью свободы.  

$$
\text{P-value} = \int_{0{,}0629}^{+\infty} f_{\chi^2(1)}(t)\,dt,
\quad\text{где}\quad
f_{\chi^2(1)}(t) = \frac{1}{\sqrt{2\pi t}}\, e^{-t/2},\; t>0.
$$

Этот интеграл можно выразить через функцию стандартного нормального распределения $\Phi$:
$$
\int_{x}^{+\infty} f_{\chi^2(1)}(t)\,dt = 2\bigl(1 - \Phi(\sqrt{x})\bigr).
$$

Подставляем $x = 0{,}0629$, $\sqrt{x} \approx 0{,}2508$.  
Значение $\Phi(0{,}2508)$ находим по таблице или приближённой формуле:

$$
\Phi(z) \approx \frac12 + \frac{1}{\sqrt{2\pi}}\left(z - \frac{z^3}{6} + \frac{z^5}{40} - \dots\right).
$$

Для $z = 0{,}2508$:
$$
\frac{z}{\sqrt{2\pi}} \approx \frac{0{,}2508}{2{,}5066} \approx 0{,}1001,
\quad \frac{z^3}{6} \approx \frac{0{,}01578}{6} \approx 0{,}00263,
$$
следующие члены малы. Тогда
$$
\Phi(0{,}2508) \approx 0{,}5 + 0{,}1001 - 0{,}00263 = 0{,}59747.
$$

Следовательно,
$$
\text{P-value} = 2 \cdot (1 - 0{,}59747) = 2 \cdot 0{,}40253 = 0{,}80506.
$$

 $P(\chi^2(1) \ge 0{,}0629) \approx 0{,}802$, что значительно больше уровня значимости $\alpha=0{,}05$.

**Вывод:** на уровне значимости 0,05 нет оснований отвергать гипотезу о том, что число смертей от удара копытом в одном корпусе за год распределено по закону Пуассона.
